```
from ultralytics import YOLO

model = YOLO("yolo11m.pt")

results=model("/content/video.mp4",save=True)
```

# To Detect only Person Class, I have set the classes Paramtere to Zero.

```
results= model ("/content/video.mp4",classes=0,save=True)
```

# Track

```
results=model.track(source="/content/video.mp4",persist=True,save=True)
```

```
results=model.track("/content/video.mp4",persist=True,show=True,classes=0, tracker="bytetrack.yaml")
```
